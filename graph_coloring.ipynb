```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def is_safe(node, color, graph, c):
    for neighbor in graph.neighbors(node):
        if color[neighbor] == c:
            return False
    return True
```

```python
def graph_coloring(graph, num_colors):
    num_nodes = len(graph.nodes)
    color = [-1] * num_nodes
    
    def color_util(node):
        if node == num_nodes:
            return True
        
        for c in range(num_colors):
            if is_safe(node, color, graph, c):
                color[node] = c
                if color_util(node + 1):
                    return True
                color[node] = -1
                
        return False
    
    if not color_util(0):
        print("Solution does not exist.")
        return None
    return color
```

```python
def visualize_graph(graph, color):
    pos = nx.spring_layout(graph)
    node_colors = [color[node] for node in graph.nodes()]
    
    nx.draw(graph, pos, with_labels=True, node_color=node_colors, cmap=plt.cm.rainbow)
    plt.show()
```

```python
if __name__ == "__main__":
    G=nx.Graph()
    G.add_edges_from([(0,1),(1,2),(2,3),(3,0),(0,2)])
    num_colors = 3
    color_solution = graph_coloring(G, num_colors)
    
    if color_solution:
        print("Graph Coloring Solution: ", color_solution)
        visualize_graph(G, color_solution)
```
